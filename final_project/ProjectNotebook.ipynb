PROJECT DESCRIPTION
    I created a chatbot for this project. The chatbot we learned in A3 is mainly continue with a free-flow conversation. I wanted to try something different, so I decided to make a chatbot that will lead the conversation toward a certain goal. The conversation will appear to be less natural but more functional. The conversation is intended to go certain way, and it should be similar to a interactive menu. By providing options, I limit the user's input so that I can set conditions that'll make the chatbot respond differently based on the input. 
   
   The goal of this chatbot is to help the user decide on what to cook. There's many instances when I wanted to cook but I couldn't on what exactly to cook. This chatbot can help by identifying preferences and use randomization to fill in other parts. This chatbot will first take in the user's dietary suggestion and filter the potential recommendations. For example, eliminate meat if user's vegan. Next, it'll take in meat preference or provide options. Then, depending on the cusine preference, this chatbot will provide spice recommendation by refering to a cooking website. Lastly, it'll randomly provide recommendations for carbs and veggies. 
   
   Aside from my lack of experience in coding, there are a few challengse I face in this project. First, I have to figure out how to keep track of the progress in this conversation and how to seperate different conditions. I ended up using running_count to execute this, since I can't really think of a better way. It works now but it can overcomplicate things if I keep expanding the code. It also makes the code more lengthy. Secondly, I have to consider how to pull out the keywords from each user's input so that the chatbot can respond correspondingly. This is different from A3 because the user's input will determine the prompt for next input. If the user's input is not what is expected, the conversation cannot progress. Thirdly, I learn the basics of BeautifulSoup in order to pull out info from a cooking website. I use indexing to selectively present the content of the website because I can't differentiate the context using just tags. There should be a more efficient way to do it but it seems to involved a lot more extended knowledge. Overall, I managed to execute what I'm hoping for with my current programming ability.

```python
import string
import random
from my_modules import module1  as mm
   
        
def get_cooking_idea():
    '''A chatbot that helps the user to decide what to cook by taking in preferences or giving suggestion
        the conversation goes in specific order described below:
        dietary restrictions, choice of protein, cuisine preference, and suggestions of carb & veggies.
    '''

    
    #this chatbot uses running_count to track progress       
    running_count=1  
    
    #begin by asking for user's dietary restriction
    msg=input("Hi! I am here to help you generate a cooking idea. "+
                    "Before we start, do you have any dietary restrctions? "+ 
                    "Are you vegan/gluten-free/both?")
    
    
    
    #specify conditions relevant to dietary restriction 
    while running_count==1:   

        #check if user want to end chatbot
        if 'quit' in msg:
            return

        
        #check if user want to refer to the list of dietary restrictions
        elif "see list" in msg:

            #list dietary restrictions and take user's dietary restriction
            mm.show_options(mm.DIETARY_RESTRICTIONS)
            msg=input('Do you have any dietary restriction described above?')
            
            
        else:
            #check if user matches any of the dietary restrictions stored or if user has none
            msg=mm.remove_punctuation(msg)
            check_dietary_restrictions=mm.check_if_in_list(msg, mm.dietary_restrictions)
            check_no_restriction=mm.check_if_in_list(msg, mm.neg_responses)

            
            #if user's response don't make sense
            if check_dietary_restrictions['in_list']==False and check_no_restriction['in_list']==False:
                 msg=input("oops, I don't recognize your dietary restriction. "+
                                "Enter \'see list\' to view dietary restrictions or \'no\' if you have none.")
    
    
            #if dietary restriction is confirmed, use running_count to mark that user is ready to move on
            #ask the user about preferred protein
            else:
                running_count=running_count+1
                msg=input("Got it! Now moving on to protein! "+
                               "Enter \'see options\' to get the lists or tell me your choice of protein: ")
                
                
                #create local variables
                proteins=mm.proteins
                carbs=mm.carbs
                
                #index local variables based on dietary restriction, remove items that cause intolerence
                try:
                    if check_dietary_restrictions ['item']=='vegan':
                        proteins=proteins[-2:] 
                        
                    if check_dietary_restrictions ['item']=='gluten-free':
                        carbs=carbs[-4:]   
                        
                    if  check_dietary_restrictions ['item']=='both':
                        proteins=proteins[-2:] 
                        carbs=carbs[-4:]  
               
                #for user that does not have dietary restriction
                except KeyError:
                    pass  
                
                

    #specify conditions relevant to protein preference
    while running_count==2:
        
            #check if user want to end chatbot
            if 'quit' in msg:
                return

            
            #check if user want to refer to the list of proteins
            #provide list and ask for user's preference
            if 'see options' in msg:
                mm.show_options(proteins)
                msg=input('Which protein do you prefer?')


            else:
                #check if user's preferred protein matches any protein in local variable 
                msg=mm.remove_punctuation(msg)
                check_proteins=mm.check_if_in_list(msg, proteins)

                
                #if preferred protein match, use running_count to indicate user is ready to move on
                #ask for cuisine preference
                if check_proteins['in_list']:
                    running_count=running_count+1
                    selected_protein=check_proteins['item']
                    msg=input("I like "+selected_protein+", too! Good choice! "+
                                    "What type of cusine are you thinking about? "+
                                    "Enter \'see options\' to see the types of cuisine I think of.")

                else:
                    msg=input("Hmm... I'm not sure about that one. "+
                                    "Enter \'see options\' to view proteins that I am familiar with.")

                    
                    
   #specify conditions relevant to cuisine preference                 
    while running_count==3:
        
            #check if user want to end chatbot
            if 'quit' in msg:
                return

            #create local variable
            cuisines=mm.cuisines
            
            #check if user want to refer to the list of cuisines
            #provide list and ask for user's preference
            if 'see options' in msg:
                mm.show_options(cuisines)
                msg=input('Choose one!')

                
            #check if user's preferred cuisine matches any cuisine in local variable 
            else:
                msg=mm.remove_punctuation(msg)
                check_cuisines=mm.check_if_in_list(msg, cuisines)

                #if preferred protein match, use running_count to indicate user is ready to move on
                #ask for cuisine preference
                if check_cuisines['in_list']:
                    running_count=running_count+1
                    selected_cuisine=check_cuisines['item']
                    msg=input(selected_cuisine+" style "+selected_protein+" it is! "+
                                    "Wanna know what what spices you can use to season it?")

                    
                else:
                    msg=input("Hmm... I'm not sure about that one. "+
                                    "Enter \'see options\' to view the cuisines that I am familiar with.")

                    
                    
    #specify conditions relevant to spices recommendations        
    while running_count==4:
            
            #check if user want to end chatbot
            if 'quit' in msg:
                return            

            
            #check if user want spice recommendation
            #if not, ask if user want carb and veggies recoomendations
            msg=mm.remove_punctuation(msg)
            recommend_spice=mm.check_if_in_list(msg, mm.neg_responses)
            if recommend_spice['in_list']:
                running_count=running_count+1
                msg=input("Alright. Last but not least, "+
                                "do you want some recoomendations on what to pair it with?")

                
            #if user want spice recommendation    
            #print a list of spices specified to the cuisine 
            else:
                running_count=running_count+1
                list_of_spices=mm.find_spice(selected_cuisine)
                print (list_of_spices)
                msg=input('Last but not least, do you want some recoomendations on what to pair it with?')


                
    #specify conditions relevant to carb and veggies recommendations
    while running_count==5:
            
            #check if user want to end chatbot
            if 'quit' in msg:
                return

            #check if user want carb and veggies recommendation
            msg=mm.remove_punctuation(msg)
            recommend_or_not=mm.check_if_in_list(msg, mm.neg_responses)
            
            
            #if user do not want recommendation
            #check if user want to continue this chatbot
            if recommend_or_not['in_list']:
                running_count=running_count+1
                print("That\'s it then. HAPPY COOKING"+"\U0001f600")

                
            #if user want recommendation  
            #seggest carb and veggies from stored variables
            #check if user want to continue this chatbot
            else:
                running_count=running_count+1
                carb_suggestion=random.choice(carbs)
                veggies_suggestion=random.choice(mm.veggies)
                print("why not try to pair it with"+carb_suggestion+". "+
                       "Throw in some "+veggies_suggestion+" , too! HAPPY COOKING"+"\U0001f600")
```
